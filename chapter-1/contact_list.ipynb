```python
data_pokemon = {
    'Name': ['Bulbasaur', 'Charmander', 'Squirtle', 'Pikachu', 'Jigglypuff'],
    'Type': ['Grass', 'Fire', 'Water', 'Electric', 'Normal'],
    'HP': [45, 39, 44, 35, 115],
    'Attack': [49, 52, 48, 55, 45],
    'Defense': [49, 43, 65, 40, 20]
}

# Simpan jadi file CSV
df_dummy = pd.DataFrame(data_pokemon)
df_dummy.to_csv('pokedex.csv', index=False)

print("File pokedex.csv berhasil dibuat! Siap dipakai.")
```

```python
pokemon = pd.read_csv('pokedex.csv')
# Tampilkan 5 baris pertama data
# pokemon.head()
# pokemon.tail()  # Tampilkan 5 baris terakhir data
# Mengambil kolom 'Name' dan 'Attack'
pokemon[['Name', 'Attack']]
# Urutkan pokemon berdasarkan serangan dari yang tertinggi ke terendah
pokemon.sort_values(by='Attack', ascending=False)
# Filter pokemon dengan elemen 'Fire'

# Buat kolom baru 'Total' yang merupakan jumlah dari HP, Attack, dan Defense
pokemon['Total'] = pokemon['HP'] + pokemon['Attack'] + pokemon['Defense']
pokemon

# Urutkan pokemon berdasarkan stat 'Total' dari yang tertinggi ke terendah
pokemon.sort_values(by='Total', ascending=False)

# Simpan DataFrame yang telah dimodifikasi ke file CSV baru
pokemon.to_csv('pokedex_modified.csv', index=False)
```
