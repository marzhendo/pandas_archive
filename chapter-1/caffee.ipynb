```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_kopi = {
    'Menu': ['Kopi Susu Gula Aren', 'Americano', 'Cappuccino', 'Matcha Latte', 'Chocolate Ice',
             'Espresso', 'Red Velvet', 'Affogato', 'Tea Tarik', 'Mineral Water'],
    'Category': ['Coffee', 'Coffee', 'Coffee', 'Non-Coffee', 'Non-Coffee',
                 'Coffee', 'Non-Coffee', 'Coffee', 'Non-Coffee', 'Non-Coffee'],
    'Price': [18000, 15000, 22000, 25000, 20000,
              12000, 23000, 28000, 15000, 5000],
    'Cups_Sold': [120, 80, 65, 50, 90,
                  40, 45, 30, 75, 100],
    'Cost_Per_Cup': [10000, 8000, 15000, 18000, 12000,
                     7000, 16000, 20000, 8000, 2000]
}

kopi_senja = pd.DataFrame(data_kopi)
kopi_senja.head()
```

```python
# Hitung Profit
# Rumus: Profit = (Price - Cost_Per_Cup) * Cups_Sold
profit = (kopi_senja['Price'] - kopi_senja['Cost_Per_Cup']) * kopi_senja['Cups_Sold']
kopi_senja['Profit'] = profit
kopi_senja.head()
```

```python
# Sortir berdasarkan Profit
kopi_senja_sorted = kopi_senja.sort_values(by='Profit', ascending=False)
kopi_senja_sorted
```

```python
# Filter menu Hemat dan Non-Coffee
hemat_non_coffee = kopi_senja[
    (kopi_senja['Price'] < 20000) & (kopi_senja['Category'] == 'Non-Coffee')
]
hemat_non_coffee
```
