```python
import pandas as pd
import numpy as np
data_gadget = {
    'Product': ['iPhone 15', 'Samsung S24', 'MacBook Air M2', 'Dell XPS 13', 'iPad Pro',
                'Galaxy Tab S9', 'Sony WH-1000XM5', 'AirPods Pro 2', 'Apple Watch 9', 'Galaxy Watch 6'],
    'Brand': ['Apple', 'Samsung', 'Apple', 'Dell', 'Apple',
              'Samsung', 'Sony', 'Apple', 'Apple', 'Samsung'],
    'Price': [16000000, 15000000, 19000000, 20000000, 14000000,
              13000000, 5000000, 4000000, 7000000, 4500000],
    'Stock': [50, 45, 20, 15, 30, 25, 60, 100, 40, 35],
    'Rating': [4.8, 4.7, 4.9, 4.5, 4.8, 4.6, 4.9, 4.7, 4.6, 4.4]
}

gadgets = pd.DataFrame(data_gadget)
gadgets.head()
```

```python
# Buat kolom baru 'Total_value' yang merupakan hasil perkalian antara 'Price' dan 'Stock'
total_value = gadgets['Price'] * gadgets['Stock']
gadgets['Total_value'] = total_value
# gadgets['Total_value'] = gadgets['Price'] * gadgets['Stock']
gadgets.head()
```

```python
# Urutkan DataFrame berdasarkan Total_value dari yang terbesar ke yang terkecil
gadgets_sorted = gadgets.sort_values(by='Total_value', ascending=False)
gadgets_sorted.head()
```

```python
# Fiter produk premium dengan kondisi Price > 15 juta dan Stock < 30
premium_gadgets = gadgets[(gadgets['Price'] > 15000000) & (gadgets['Stock'] < 30)]
premium_gadgets
```
