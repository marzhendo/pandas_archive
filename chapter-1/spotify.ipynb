```python
import pandas as pd
import numpy as np
```

```python
data_spotify = {
    'Song': ['Blinding Lights', 'Shape of You', 'Someone Like You', 'Dance Monkey', 'Rockstar',
             'As It Was', 'Sunflower', 'One Dance', 'Stay', 'Believer'],
    'Artist': ['The Weeknd', 'Ed Sheeran', 'Adele', 'Tones and I', 'Post Malone',
               'Harry Styles', 'Post Malone', 'Drake', 'Kid LAROI', 'Imagine Dragons'],
    'Genre': ['Pop', 'Pop', 'Soul', 'Pop', 'Hip-Hop',
              'Pop', 'Hip-Hop', 'Hip-Hop', 'Pop', 'Rock'],
    'Duration_Min': [3.5, 4.0, 4.8, 3.5, 3.6,
                     2.8, 2.6, 3.0, 2.3, 3.4],
    'Streams_Millions': [3800, 3600, 3000, 2900, 2700,
                         2600, 2500, 2400, 2300, 2200]
}

spotify = pd.DataFrame(data_spotify)
spotify.head()
```

```python
# Menghitung total waktu putar semua lagu dalam menit
total_minutes = spotify['Duration_Min'] * spotify['Streams_Millions']
spotify['Total_Minutes'] = total_minutes
spotify.head()
```

```python
# Urutkan lagu berdasarkan Total_Minutes dari yang tertinggi
spotify_sorted = spotify.sort_values(by='Total_Minutes', ascending=False)
spotify_sorted
```
