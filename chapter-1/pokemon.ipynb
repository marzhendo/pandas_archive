```python
import numpy as np
import pandas as pd
```

```python
data_pokemon = {
    'Name': ['Pikachu', 'Charizard', 'Mewtwo', 'Bulbasaur', 'Squirtle',
             'Gyarados', 'Dragonite', 'Onix', 'Lugia', 'Gengar'],
    'Type': ['Electric', 'Fire', 'Psychic', 'Grass', 'Water',
             'Water', 'Dragon', 'Rock', 'Psychic', 'Ghost'],
    'HP': [35, 78, 106, 45, 44,
           95, 91, 35, 106, 60],
    'Attack': [55, 84, 110, 49, 48,
               125, 134, 45, 90, 65],
    'Defense': [40, 78, 90, 49, 65,
                79, 95, 160, 130, 60],
    'Legendary': [False, False, True, False, False,
                  False, False, False, True, False]
}

pokedex = pd.DataFrame(data_pokemon)
pokedex.head()
```

```python
# Buat kolom baru 'Total_Stats' yang merupakan jumlah dari HP, Attack, dan Defense
pokedex['Total'] = pokedex['HP'] + pokedex['Attack'] + pokedex['Defense']
pokedex.head()
```

```python
# Cari pokemon terkuat berdasarkan Total_Stats
strongest_pokemon = pokedex.sort_values(by='Total', ascending=False)
strongest_pokemon
```

```python
# Filter pokemon dengan tipe 'Water' dan Attack di atas 100
water_strong_pokemon = pokedex[
    (pokedex['Type'] == 'Water') & (pokedex['Attack'] > 100)
]
water_strong_pokemon
```

```python
# Filter pokemon legendaris dengan Attack di atas 100
legendary_strong = pokedex[
    (pokedex['Legendary'] == True) & (pokedex['Attack'] > 100)
]
legendary_strong
```

```python
# Filter pokemon legendaris
legendary_pokemon = pokedex[pokedex['Legendary'] == True]
legendary_pokemon
```
